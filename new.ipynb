```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/MbuyiseloNtando/Sustainable-development-report-2000-2023-/main/sustainable_development_report_2023.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.country.duplicated().sum()
```

```python
df1 = df.head(20).groupby("country")[["goal_1_score", "goal_2_score", "goal_3_score", "goal_4_score", "goal_5_score"]].mean()
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df1.plot(kind="bar", legend=True, figsize = (10,5))
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (10,5))
axes.plot(df1)
ax=sns.barplot(data=df.head(40), x=df.country.head(40), y=df["goal_2_score"].head(40), legend=True)
plt.xticks(rotation=45, ha='right')
plt.show()
```
