```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/MbuyiseloNtando/Sustainable-development-report-2000-2023-/main/sustainable_development_report_2023.csv")
```

```python
df.head()
```

```python
df.country.duplicated().sum()
```

```python
df1 = df.head(20).groupby("country")["goal_1_score"].mean()
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(figsize = (10,5))
axes.plot(df1)
ax=sns.barplot(data=df.head(40), x=df.head(40).country, y=df.head(40).goal_1_score.dropna(), legend=False)
plt.xticks(rotation=45, ha='right')
plt.show()
```
